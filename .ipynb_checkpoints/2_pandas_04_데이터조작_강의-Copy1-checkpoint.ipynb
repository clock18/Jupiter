### ❓ 연습문제 : 수치형-> 범주형 데이터 생성

: 타이타닉호 승객을 사망자와 생존자 그룹으로 나누고(alive), 각 그룹에 대해 미성년자, 청년, 중년, 장년, 노년 승객의 비율을 구하시오.

- bins=[1,15,25,35,60,99]
- labels=['미성년자','청년','중년','장년','노년']
- 각 그룹별 비유의 전체 합은 1이 되어야 한다.

**타이타닉 데이터셋 로딩**

```python
# 타이타닉 승객 dataset 읽어오기
import seaborn as sns 
titanic = sns.load_dataset('titanic')
titanic.head()
```

**데이터 셋 분리 : 생존자 그룹과 사망자 그룹**

```python
# 사망자 그룹 데이터 프레임 추출
titanicDead = titanic[titanic['alive'] == 'no']
titanicDead
```

```python
# 생존자 그룹 데이터 프레임 추출
titanicAlive = titanic[titanic['alive'] == 'yes']
titanicAlive
```

**데이터 범위 설정**

```python
# 라벨과 범위 간격 설정

bins=[0.1,15,25,35,60,99]
labels=['미성년자','청년','중년','장년','노년']
```

```python
# 범위 유효성 검사
# bins=[0,15,25,35,60,99]
titanicDead['age'].min()
titanicDead['age'].max()

titanicAlive['age'].min()
titanicAlive['age'].max()



#제시된 bins의 최초값 1 이하인 값 1.0과 0.42가 각
#그룹의 최소나이로 확인되므로 bins 수정
#최대나이는 74와 80으로 확인되어서 수정 필요 없음
```

**범주형 데이터 생성**
