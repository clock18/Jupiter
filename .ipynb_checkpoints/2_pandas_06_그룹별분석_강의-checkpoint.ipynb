```python
data = {
    "ë„ì‹œ": ["ì„œìš¸", "ì„œìš¸", "ì„œìš¸", "ë¶€ì‚°", 
           "ë¶€ì‚°", "ë¶€ì‚°", "ì¸ì²œ", "ì¸ì²œ"],
    "ì—°ë„": ["2015", "2010", "2005", "2015",
           "2010", "2005", "2015", "2010"],
    "ì¸êµ¬": [9904312, 9631482, 9762546, 3448737,
           3393191, 3512547, 2890451, 263203],
    "ì§€ì—­": ["ìˆ˜ë„ê¶Œ", "ìˆ˜ë„ê¶Œ", "ìˆ˜ë„ê¶Œ", "ê²½ìƒê¶Œ",
           "ê²½ìƒê¶Œ", "ê²½ìƒê¶Œ", "ìˆ˜ë„ê¶Œ", "ìˆ˜ë„ê¶Œ"]
}

columns = ["ë„ì‹œ", "ì—°ë„", "ì¸êµ¬", "ì§€ì—­"]

df1 = pd.DataFrame(data, columns=columns)
df1
```

### ì˜ˆì œ2. íƒ€ì´íƒ€ë‹‰ ë°ì´í„° : pivot_table() ì‚¬ìš©

```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset('titanic')[['age','sex','class','fare','survived']]
df.head()
```

**í”¼ë´‡í…Œì´ë¸”1 : ì„ ì‹¤ë“±ê¸‰ë³„ ìˆ™ë°•ê°ì˜ ì„±ë³„ í‰ê·  ë‚˜ì´**

```python
# ê° ì„ ì‹¤ ë“±ê¸‰ë³„ë¡œ ìˆ™ë°•ê°ì˜ ì„±ë³„ í‰ê·  ë‚˜ì´
pd.pivot_table(df, index='class', columns='sex', values='age')
```

**í”¼ë´‡í…Œì´ë¸”2 : ì„ ì‹¤ë“±ê¸‰ë³„ ìˆ™ë°•ê°ì˜ ìƒì¡´ì—¬ë¶€ì— ë”°ë¥¸ í‰ê·  ë‚˜ì´**

```python
# ê° ì„ ì‹¤ ë“±ê¸‰ë³„ ìˆ™ë°•ê°ì˜ ìƒì¡´ì ìˆ˜ì™€ ìƒì¡´ìœ¨
pd.pivot_table(df, index='class', columns='survived', 
               values='age')
```

**í”¼ë´‡í…Œì´ë¸”3 : ì„ ì‹¤ë“±ê¸‰ê³¼ ì„±ë³„ì— ëŒ€í•´ ìƒì¡´ì—¬ë¶€ì— ë”°ë¥¸ ë‚˜ì´ì™€ í‹°ì¼“ê°’ì˜ í‰ê· ê³¼ ìµœëŒ€ê°’ ê³„ì‚°**

```python
pd.pivot_table(df, index=['class', 'sex'], columns='survived', 
               values=['age','fare'],aggfunc=['mean','max'])
```

### ì˜ˆì œ ë°ì´í„° ì¤€ë¹„

```python
df5 = pd.DataFrame({'key1' : ['A', 'A', 'B', 'B', 'A'],
                    'key2' : ['one', 'two', 'one', 'two', 'one'],
                    'data1' : [i for i in range(1,6)],
                    'data2' : [i for i in range(10,60,10)]})
df5
```

### ë‹¨ì¼ ì»¬ëŸ¼ì˜ ê·¸ë£¹í™”

```python
groups = df5.groupby(df5.key1)
groups
```

**groups ì†ì„± : GroupByí´ë˜ìŠ¤ê°ì²´ groupsì˜ ê·¸ë£¹ì •ë³´ í™•ì¸**
- ê° ê·¸ë£¹ì˜ ì´ë¦„ì„ í‚¤ë¡œ ê°–ëŠ” íŒŒì´ì¬ ë”•ì…”ë„ˆë¦¬ë¥¼ ë°˜í™˜
- ë”•ì…”ë„ˆë¦¬ ê°’ì€ ê° ê·¸ë£¹ì— ì†í•˜ëŠ” ì¸ë±ìŠ¤ ë ˆì´ë¸”ì˜ ë°°ì—´
- Dict {group name -> group labels}

```python
groups.groups   # ì†í•œ ì¸ë±ìŠ¤ ë²ˆí˜¸ ë„ì¶œ
```

**ngroups ì†ì„± : group ìˆ˜ ë°˜í™˜**

```python
groups.ngroups 
```

**ê·¸ë£¹í™”ëœ ê²°ê³¼ ì ‘ê·¼**
- ê·¸ë£¹ ë‚´ìš© ì¶œë ¥í•¨ìˆ˜ ì •ì˜ : print_groups(group_object)
